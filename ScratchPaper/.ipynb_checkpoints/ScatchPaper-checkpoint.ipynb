# Instances Data

## Ins 1

```python
csv_data_1 = pd.read_csv('/Users/zzq/Desktop/ERP_2021/HUAWEI_dpdp/code/Git/simulator/dpdp_competition/benchmark/instance_1/50_1.csv')
csv_data_1
```

```python

def plot_orders(csv_data, order_number, just_this_order = True):

    plt.plot(x,y,'.', color='silver')
    
    plt.axis('equal')
    
    for i in range(len(csv_data['order_id'])):
        
        if i < order_number and just_this_order:
            continue
            
            
        pick_loc = (f_dict[csv_data['pickup_id'][i]])
        
        dili_loc = (f_dict[csv_data['delivery_id'][i]])
        
        plt.annotate(str(i), xy=pick_loc, xytext=pick_loc, color = 'r')
        plt.annotate(str(i), xy=dili_loc, xytext=dili_loc, color = 'b')
        
        
        
        if i == order_number:
            break
        
    fig = plt.gcf()
    fig.set_size_inches(10, 10)
    
    return
```

```python
plot_orders(csv_data_1, 50, False)
```

```python
def calculate_location_conges(csv_data, node = 'pickup_id'):
    dic = {}
    for i in range(len(csv_data['order_id'])):
        node_id = csv_data[node][i]
        if node_id not in dic:
            dic[node_id] = 1
        else:
            dic[node_id] += 1
            
    return dic
```

```python
dic_1 = calculate_location_conges(csv_data_1)
```

```python
dic_1
```

```python
x = list(dic_1.keys())
y = list(dic_1.values())
fig, ax = plt.subplots(figsize=(10, 7))
ax.bar(x=x, height=y, color = 'r')
ax.set_title("Distribution of pickup orders", fontsize=15)
plt.xticks([])
```

```python
dic_d_1 = calculate_location_conges(csv_data_1, 'delivery_id')
```

```python
dic_d_1
```

```python
x_d = list(dic_d_1.keys())
y_d = list(dic_d_1.values())
fig, ax = plt.subplots(figsize=(10, 7))
ax.bar(x=x_d, height=y_d, color = 'b')
ax.set_title("Distribution of delivery orders", fontsize=15)
plt.xticks([])
```

```python
shops = x + [i for i in x_d if i not in x]

y1 = []
for i in range(len(shops)):
    if shops[i] not in dic_1:
        y1.append(0)
    else:
        y1.append(dic_1[shops[i]])
        
y2 = []
for i in range(len(shops)):
    if shops[i] not in dic_d_1:
        y2.append(0)
    else:
        y2.append(dic_d_1[shops[i]])

fig, ax = plt.subplots(figsize=(10, 7))
# 先创建一根柱子，显示第一种产品的销量
ax.bar(shops, y1, color="red", label="pickup")
# 第二根柱子“堆积”在第一根柱子上方，通过'bottom'调整，显示第二种产品的销量
ax.bar(shops, y2, color="blue", bottom=y1, label="delivery")


ax.set_title("Distribution of orders", fontsize=15)
ax.set_xlabel("Nodes")
ax.set_ylabel("Number of orders")
ax.legend()
plt.xticks([])
```
